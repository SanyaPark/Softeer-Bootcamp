```python
import requests
from bs4 import BeautifulSoup# 크롤링
import pandas as pd # 데이터 처리
import json # to json file
```

```python
url = 'https://www.imf.org/external/datamapper/region.htm#sea'
```

```python
response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
else:
    print(response.status_code)
    
```

```python
table = soup.select('div.fancy > table > tr')
table
```

```python
ctry = [[td.get_text(strip=False) for td in tr.find_all('th')] for tr in table]
final_result = []
for element in table:
    if element.p:
        # <p> 태그 내부의 내용을 줄 단위로 분리
        lines = element.p.decode_contents().split('<br/>')
        
        # 각 라인을 처리: 중간 <br/> 무시, 끝의 <br/>만 고려
        processed_lines = []
        for line in lines:
            cleaned_line = line.replace('<br/>', '').strip()  # <br/> 제거
            if cleaned_line:  # 내용이 있다면 추가
                processed_lines.append(cleaned_line)
        
        # 한 문장으로 합치고 최종 결과에 추가
        final_result.extend(processed_lines)
```
