```python
import requests
from bs4 import BeautifulSoup# 크롤링
import pandas as pd # 데이터 처리
import json # to json file
```

```python
nameURL = 'https://www.imf.org/external/datamapper/api/v1/countries'
gdpURL = 'https://www.imf.org/external/datamapper/api/v1/NGDPD'
```

```python
response = requests.get(nameURL)
names = response.text
names
```

```python
namedict = json.loads(names)
```

```python
namedict = namedict['countries']
namedict
```

```python
rspn = requests.get(gdpURL)
gdptext = rspn.text
```

```python
gdpdict = json.loads(gdptext)
gdpdict = gdpdict['values']['NGDPD']
gdpdict
```

```python
# 이 gdpdict에 국가명이 없으면 IMF의 집계 대상이 아닌 것
gdps = {}
for k, v in gdpdict.items():
    
    if k in namedict.keys():
        print(k)
        if '2025' in v.keys():
            gdps[namedict[k]['label']] = v['2025']
        else:
            gdps[namedict[k]['label']] = -1
            
```

```python
len(gdps)
gdps = {k:v for k, v in gdps.items() if v >= 100}
```
